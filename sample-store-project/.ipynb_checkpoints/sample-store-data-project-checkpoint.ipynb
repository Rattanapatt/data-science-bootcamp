```python
# TODO 03 - your friend ask for `California` data, filter it and export csv for him
filtered_df = df[df['State'] == 'California'].reset_index()
filtered_df.rename(columns = {'index' : 'Original Index'}, inplace = True)
filtered_df.to_csv('California_Customer_data.csv')
# filtered_df['State'].value_counts()
filtered_df
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
import numpy as np
data_2017 = df[df['Order Date'].dt.year == 2017].reset_index().drop('index', axis = 1)
sale_summary = data_2017['Sales'].agg(['sum', 'mean', 'std']).rename(index = {'sum' : 'Total sales in 2017',
                                                               'mean' :'Average sales in 2017',
                                                               'std' : 'Standard Deviation of Sales in 2017'}
                                                            ).round(2)
sale_summary.rename('Sales in USD($)', inplace=True)
[print(f'{variable} is {sale}.') for variable, sale in zip(sale_summary.index, sale_summary.values)]

sale_summary
```

```python
# TODO 06 - which Segment has the highest profit in 2018
profit_seg_2018 = df[df['Order Date'].dt.year == 2017].reset_index().drop('index', axis = 1).groupby('Segment').Profit.sum().round(2)
segment_2018 = profit_seg_2018.idxmax()
max_profit_2018 = profit_seg_2018.max()
print(f'The segment that has the highest profit is {segment_2018} with {max_profit_2018}$.')
profit_seg_2018
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
filtered_sales = df[(df['Order Date'] >= '2019-04-15') & (df['Order Date'] <= '2019-12-31')]
filtered_sales = filtered_sales.groupby('State').Sales.sum()
smallest5_sales = filtered_sales.nsmallest(5)
print('The 5 states that have the lowest total sales between 15 April 2019 - 31 December 2019 are...')
[print(f'{state}, with sales amounting to {sale}$') for state, sale in zip(smallest5_sales.index, smallest5_sales.values)]

smallest5_sales
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25%
sales_2019 = df[df['Order Date'].dt.year == 2019].reset_index().drop('index', axis = 1).groupby('Region')['Sales'].sum().round(2)
CW_sales = sales_2019.Central + sales_2019.West
total_sales = sales_2019.sum()
proportion = (CW_sales/total_sales * 100).round(2)

print(f'The proportion of sales from West combined with Central in 2019 is {proportion}%.')

sales_2019
#df.Region.unique()
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
start_date = 2019
end_date = 2020
filtered_data = df[df['Order Date'].dt.year.isin([start_date, end_date])].reset_index().drop(['index', 'Row ID'] , axis = 1)
by_order = filtered_data['Product Name'].value_counts().nlargest(10)
by_sales = filtered_data.groupby('Product Name')['Sales'].sum().round(2).nlargest(10)

print(f'Top 10 most order products are\n...')
[print(f'{prd} with orders of {quant} units.') for prd, quant in zip(by_order.index, by_order.values)]

print('\n--------------------------------------------------------------------------------------------------------------------\n')

print(f'Top 10 most sales products are\n...')
[print(f'{prd} with sales amounting of {sales}$.') for prd, sales in zip(by_sales.index, by_sales.values)]

print('\n--------------------------------------------------------------------------------------------------------------------\n')

print(by_order)
print(by_sales)
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
product_sales = df.groupby('Product Name')[['Sales', 'Profit']].agg('sum').nlargest(10, 'Profit')
product_sales.plot(kind = 'bar', color = ['red', 'green'])

product_sales = df.groupby('Product Name')[['Sales', 'Profit']].agg('sum').nsmallest(10, 'Profit')
product_sales.plot(kind = 'bar', color = ['red', 'green'])
```

```python
# TODO Bonus - use np.where() to create new column in dataframe to help you answer your own questions
# I wonder if we sell something and we loss there were no reason to sell.
# On the other hand minimal number of %profit also means ineffective profit generation from the product which should also avoid.
import numpy as np
import matplotlib.pyplot as plt

product_sales = df.groupby('Product Name')[['Sales', 'Profit']].agg('sum')
product_sales['%Profit'] = (product_sales['Profit']/product_sales['Sales']*100).round(2)
product_sales[['Sales', 'Profit']] = product_sales[['Sales', 'Profit']].round(2)
product_sales['Good Product'] = np.where(product_sales['%Profit'] > 30, True, False)
product_sales[product_sales['Good Product'] == True].reset_index()

good_products = product_sales[product_sales['Good Product']].reset_index()
ngood_products = len(product_sales[product_sales['Good Product'] == True].reset_index())
bad_products = product_sales[~product_sales['Good Product']].reset_index()
nbad_products = len(product_sales[~product_sales['Good Product'] == True].reset_index())

print(f'There were {ngood_products} products that yielded an acceptable profit, each generating more than a 30% profit from sales.')
good_products.sort_values('%Profit', ascending = False).reset_index().drop('index', axis = 1)

print(f'There were {nbad_products} products that yielded an unacceptable profit, each generating less than or euqal to a 30% profit from sales.')
good_products.sort_values('%Profit', ascending = True).reset_index().drop('index', axis = 1)

good_products.plot(x = 'Sales', y = 'Profit', kind = 'scatter', color='green', label='Good Products', s = 5, alpha = 0.4)
bad_products.plot(x = 'Sales', y = 'Profit', kind = 'scatter', color='red', label='Good Products', s = 5, alpha = 0.4)

[fig, ax] = plt.subplots()

ax.scatter(good_products['Sales'], good_products['Profit'], color='green', label='Good Products', s = 5, alpha = 0.01)
ax.scatter(bad_products['Sales'], bad_products['Profit'], color='red', label='Bad Products', s = 5, alpha= 0.01)

ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
ax.set_title('Sales vs Profit Comparison')

ax.legend()
plt.show()
```
